```python
from tkinter import *
import tkinter as tk
from tkinter import *
from tkinter import messagebox, filedialog
from Crypto.Cipher import Blowfish
from PIL import Image, ImageTk
import base64
import os

import random
#from colorthief import ColorThief
from tkinter import filedialog as fd
from stegano import lsb
import pyttsx3
```

```python
keyE = str(input("Enter the E - Key: "))
keyD = str(input("Enter the D - Key: "))
```

```python
def encrypt():
    password = code.get()
    
    if password==keyE:
        key = code.get().encode()
        message=text1.get(1.0,END).encode()
        cipher = Blowfish.new(key, Blowfish.MODE_ECB)
        encrypted_data = cipher.encrypt(pad(message, Blowfish.block_size)) 
        
        f1.insert(END, encrypted_data.hex(),'encrt')
        f1.insert(END,"\n....................Encrypted successfully .../\n",'encrt')
        
    elif password=="":
        messagebox.showerror("encryption","Input Password")
        f1.insert(END,"\n _encryption key is unavailable (empty).../\n")
    elif password !=keyE:
        messagebox.showerror("encryption","INVALID Password")
        messagebox.showerror("encryption","Encryption Denied")
        f1.insert(END,"\n _encryption key is incorrect (WARNING).../\n")
        
def decrypt():
    password=code.get()
                            
    if password==keyD:
        key = code.get().encode()
        ciphertext_hex = text1.get(1.0, END).strip()
        ciphertext = bytes.fromhex(ciphertext_hex)
        
        cipher = Blowfish.new(key, Blowfish.MODE_ECB)
        decrypted_data = unpad(cipher.decrypt(ciphertext), Blowfish.block_size)                         
                                 
        f1.insert(END, decrypted_data.decode(),'decrp')
        f1.insert(END,"\n....................Decrypted successfully .../\n",'decrp')
        
    elif password=="":
        messagebox.showerror("decryption","Input Password")
        f1.insert(END,"\n _decryption key is unavailable (empty).../\n")    
    
    elif password !=keyD:
        messagebox.showerror("decryption","INVALID Password")
        messagebox.showerror("decryption","Decryption Denied")
        f1.insert(END,"\n _decryption key is incorrect (WARNING).../\n")
                                 
def pad(data, block_size):
    padding_length = block_size - len(data) % block_size
    padding = bytes([padding_length]) * padding_length
    return data + padding

def unpad(data, block_size):
    padding_length = data[-1]
    if padding_length > block_size:
        raise ValueError("Invalid padding")
    return data[:-padding_length]                                 
        
```
